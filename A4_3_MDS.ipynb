```python
import pandas as pd
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
```

```python
icecream_df=pd.read_csv('/content/icecream.csv')
icecream_df
```

```python
ice = icecream_df.drop(columns=['Brand'])
```

```python
distance_matrix = ice.to_numpy()

# Perform MDS with 2 dimensions
mds = MDS(n_components=2, dissimilarity='euclidean')
mds_result = mds.fit_transform(distance_matrix)
```

```python
# Plotting the MDS results
plt.figure(figsize=(10, 8))
plt.scatter(mds_result[:, 0], mds_result[:, 1], marker='o', color='blue')
# Adding labels for each point (brand)
for i, brand in enumerate(icecream_df['Brand']):
    plt.text(mds_result[i, 0], mds_result[i, 1], brand, fontsize=9)

plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('MDS Plot for Ice Cream Brands')
plt.grid(True)
plt.show()
```
