```python
import pandas as pd
from statsmodels.api import OLS
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
```

```python
pizza=pd.read_csv("/content/pizza_data.csv")
pizza.head()
```

```python
pizza.info()
```

```python
label_encoder = LabelEncoder()
for col in pizza.columns:
    if pizza[col].dtype == 'object':  # Check if the column contains strings
        pizza[col] = label_encoder.fit_transform(pizza[col])
```

```python
x=pizza.drop(['ranking'],axis=1).values
y=pizza['ranking'].values
```

```python
model = sm.OLS(y, sm.add_constant(x))
results = model.fit()
```

```python
print(results.summary())
```
